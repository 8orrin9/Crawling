```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import NoSuchElementException, ElementClickInterceptedException
from selenium.common.exceptions import StaleElementReferenceException
import time
import os
from urllib.request import urlretrieve
import re
import pandas as pd
```

```python
# 웹 드라이버 실행
driver = wb.Chrome()
```

```python
# 페이지 스크롤 및 댓글 수집
previous_height = driver.execute_script("return document.body.scrollHeight")
comments_list = []
```

```python
while True:
    try:
        # 댓글 요소를 다시 찾기
        comments = driver.find_elements(By.CSS_SELECTOR, "div.css-175oi2r[data-testid='tweetText']") #css-175oi2r means div class
        for comment in comments:
            try:
                text = comment.text.strip()
                if text:  # 댓글이 비어 있지 않은 경우만 저장
                    comments_list.append(text)
            except StaleElementReferenceException:
                # 요소가 더 이상 존재하지 않을 경우 무시
                continue
        
        # 스크롤을 페이지 하단으로 내리기
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)  # 로드 대기 시간 (페이지에 따라 조정 가능)
        
        # 새 페이지 높이 확인
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == previous_height:
            # 페이지 끝에 도달한 경우 루프 종료
            break
        previous_height = new_height
    
    except StaleElementReferenceException:
        # 페이지 또는 요소가 새로 고쳐진 경우
        print("페이지가 업데이트되었습니다. 다시 시도합니다.")
        time.sleep(2)  # 잠시 대기 후 다시 시도
```

```python
comments = pd.DataFrame(comments_list)

# 행 길이 최대화
pd.set_option('display.max_colwidth', None)
comments.head(30)
```
